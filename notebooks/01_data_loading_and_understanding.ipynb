This notebook loads the OULAD CSV files, inspects their structure, identifies key columns, and prepares the first base dataset for the Potential Pathway Index project.

The goal of this notebook is not to build a model yet.  
The goal is to understand the raw data before making any transformation.

```python
from pathlib import Path

import pandas as pd
import numpy as np
```

```python
PROJECT_ROOT = Path("..")

DATA_RAW = PROJECT_ROOT / "data" / "raw"
DATA_PROCESSED = PROJECT_ROOT / "data" / "processed"

DATA_PROCESSED.mkdir(parents=True, exist_ok=True)

print("Project root:", PROJECT_ROOT.resolve())
print("Raw data path:", DATA_RAW.resolve())
print("Processed data path:", DATA_PROCESSED.resolve())
```

```python
expected_files = {
    "assessments": "assessments.csv",
    "courses": "courses.csv",
    "student_assessment": "studentAssessment.csv",
    "student_info": "studentInfo.csv",
    "student_registration": "studentRegistration.csv",
    "student_vle": "studentVle.csv",
    "vle": "vle.csv"
}

for table_name, file_name in expected_files.items():
    file_path = DATA_RAW / file_name
    print(f"{table_name:25} | exists: {file_path.exists()} | file: {file_name}")
```

```python
dataframes = {}

for table_name, file_name in expected_files.items():
    file_path = DATA_RAW / file_name
    
    if file_path.exists():
        dataframes[table_name] = pd.read_csv(file_path)
        print(f"{table_name:25} | loaded successfully | shape: {dataframes[table_name].shape}")
    else:
        print(f"{table_name:25} | file not found | expected file: {file_name}")
```

```python
assessments = dataframes["assessments"]
courses = dataframes["courses"]
student_assessment = dataframes["student_assessment"]
student_info = dataframes["student_info"]
student_registration = dataframes["student_registration"]
student_vle = dataframes["student_vle"]
vle = dataframes["vle"]

print("All tables were assigned successfully.")
```

```python
def inspect_table(df, table_name):
    """
    Inspect the basic structure of a DataFrame.

    This function shows:
    - number of rows and columns
    - column names
    - data types
    - missing values
    - first rows
    """
    
    print("=" * 80)
    print(f"TABLE: {table_name}")
    print("=" * 80)
    
    print("\nShape:")
    print(df.shape)
    
    print("\nColumns:")
    print(df.columns.tolist())
    
    print("\nData types:")
    print(df.dtypes)
    
    print("\nMissing values:")
    missing_values = df.isnull().sum()
    print(missing_values[missing_values > 0].sort_values(ascending=False))
    
    print("\nFirst rows:")
    display(df.head())
```

```python
inspect_table(assessments, "assessments")
inspect_table(courses, "courses")
inspect_table(student_assessment, "studentAssessment")
inspect_table(student_info, "studentInfo")
inspect_table(student_registration, "studentRegistration")
inspect_table(student_vle, "studentVle")
inspect_table(vle, "vle")
```

```python
main_keys = ["id_student", "code_module", "code_presentation"]

print("Main key columns:")
print(main_keys)
```

```python
student_info_total_rows = student_info.shape[0]
student_info_unique_keys = student_info[main_keys].drop_duplicates().shape[0]

print("student_info total rows:", student_info_total_rows)
print("student_info unique student-module-presentation keys:", student_info_unique_keys)
print("Is the main key unique in student_info?", student_info_total_rows == student_info_unique_keys)
```

```python
student_registration_total_rows = student_registration.shape[0]
student_registration_unique_keys = student_registration[main_keys].drop_duplicates().shape[0]

print("student_registration total rows:", student_registration_total_rows)
print("student_registration unique student-module-presentation keys:", student_registration_unique_keys)
print("Is the main key unique in student_registration?", student_registration_total_rows == student_registration_unique_keys)
```

```python
student_info_keys = student_info[main_keys].drop_duplicates()
student_registration_keys = student_registration[main_keys].drop_duplicates()

merged_key_check = student_info_keys.merge(
    student_registration_keys,
    on=main_keys,
    how="outer",
    indicator=True
)

print("Key match status between student_info and student_registration:")
print(merged_key_check["_merge"].value_counts())
```

```python
assessments_total_rows = assessments.shape[0]
assessments_unique_ids = assessments["id_assessment"].nunique()

print("assessments total rows:", assessments_total_rows)
print("unique id_assessment values:", assessments_unique_ids)
print("Is id_assessment unique in assessments?", assessments_total_rows == assessments_unique_ids)
```

```python
student_assessment_ids = student_assessment["id_assessment"].drop_duplicates().to_frame()
assessment_ids = assessments[["id_assessment"]].drop_duplicates()

assessment_coverage_check = student_assessment_ids.merge(
    assessment_ids,
    on="id_assessment",
    how="left",
    indicator=True
)

print("Assessment coverage between student_assessment and assessments:")
print(assessment_coverage_check["_merge"].value_counts())
```

```python
student_vle_site_ids = student_vle["id_site"].drop_duplicates().to_frame()
vle_site_ids = vle[["id_site"]].drop_duplicates()

vle_coverage_check = student_vle_site_ids.merge(
    vle_site_ids,
    on="id_site",
    how="left",
    indicator=True
)

print("VLE site coverage between student_vle and vle:")
print(vle_coverage_check["_merge"].value_counts())
```

```python
key_validation_summary = {
    "student_info_main_key_unique": student_info_total_rows == student_info_unique_keys,
    "student_registration_main_key_unique": student_registration_total_rows == student_registration_unique_keys,
    "assessment_id_unique": assessments_total_rows == assessments_unique_ids,
    "student_info_registration_key_match": merged_key_check["_merge"].value_counts().to_dict(),
    "student_assessment_coverage": assessment_coverage_check["_merge"].value_counts().to_dict(),
    "vle_site_coverage": vle_coverage_check["_merge"].value_counts().to_dict()
}

key_validation_summary
```

```python
final_result_values = student_info["final_result"].value_counts()

print("final_result distribution:")
print(final_result_values)
```

```python
student_base = student_info.copy()

print("student_base created successfully.")
print("Shape:", student_base.shape)
```

```python
risk_results = ["Withdrawn", "Fail"]
stable_results = ["Pass", "Distinction"]

student_base["at_risk_misalignment"] = np.where(
    student_base["final_result"].isin(risk_results),
    1,
    0
)

student_base[
    ["id_student", "code_module", "code_presentation", "final_result", "at_risk_misalignment"]
].head(10)
```

```python
target_mapping_check = pd.crosstab(
    student_base["final_result"],
    student_base["at_risk_misalignment"]
)

target_mapping_check
```

```python
target_counts = student_base["at_risk_misalignment"].value_counts()
target_percentages = student_base["at_risk_misalignment"].value_counts(normalize=True) * 100

target_summary = pd.DataFrame({
    "count": target_counts,
    "percentage": target_percentages.round(2)
})

target_summary
```

```python
safe_registration_features = student_registration[
    main_keys + ["date_registration"]
].copy()

student_base = student_base.merge(
    safe_registration_features,
    on=main_keys,
    how="left"
)

print("student_base after adding safe registration features:")
print(student_base.shape)

student_base.head()
```

```python
registration_missing = student_base["date_registration"].isnull().sum()

print("Missing date_registration values after merge:", registration_missing)
print("Total rows in student_base:", student_base.shape[0])
```

```python
base_columns = [
    "id_student",
    "code_module",
    "code_presentation",
    "gender",
    "region",
    "highest_education",
    "imd_band",
    "age_band",
    "num_of_prev_attempts",
    "studied_credits",
    "disability",
    "date_registration",
    "final_result",
    "at_risk_misalignment"
]

base_student_dataset = student_base[base_columns].copy()

print("Initial base dataset created.")
print("Shape:", base_student_dataset.shape)

base_student_dataset.head()
```

```python
print("Shape:")
print(base_student_dataset.shape)

print("\nColumns:")
print(base_student_dataset.columns.tolist())

print("\nMissing values:")
print(base_student_dataset.isnull().sum().sort_values(ascending=False))

print("\nTarget distribution:")
print(base_student_dataset["at_risk_misalignment"].value_counts(normalize=True).round(4))
```

```python
base_dataset_path = DATA_PROCESSED / "base_student_dataset.csv"

print("Base dataset output path:")
print(base_dataset_path)
```

```python
base_student_dataset.to_csv(base_dataset_path, index=False)

print("Base student dataset saved successfully.")
print("Saved file:", base_dataset_path)
print("Shape:", base_student_dataset.shape)
```

```python
base_student_dataset_check = pd.read_csv(base_dataset_path)

print("Reloaded dataset shape:")
print(base_student_dataset_check.shape)

print("\nReloaded dataset columns:")
print(base_student_dataset_check.columns.tolist())

print("\nFirst rows:")
display(base_student_dataset_check.head())
```
